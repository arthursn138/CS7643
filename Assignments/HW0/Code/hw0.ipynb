#### Q14: Without using `np.pad`, pad an array with a border of zeros

```python
x = np.ones((4, 4))
print(x)
# print(np.pad(x, 1))
# print(hw0.q14(x))


r, c = x.shape
v = np.zeros((c))
h = np.zeros((r+2))
# print(v, h.T)
# x = np.hstack((h.T, x, h.T))
x = np.vstack((v, x, v))
print(x)
x = np.concatenate((h.T, x, h.T))
print(h, x)
```
